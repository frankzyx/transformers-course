# Byte-Pair Encoding  (BPE) Tokenization

Implementation

````python
corpus = [
    "This is the Hugging Face Course.",
    "This chapter is about tokenization.",
    "This section shows several tokenizer algorithms.",
    "Hopefully, you will be able to understand how they are trained and generate tokens.",
]
````

````python
# pre-tokenization
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('gpt2')
````

````python
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str(corpus[0])
````

````python
from collections import defaultdict

word_freqs = defaultdict(int)

for text in corpus:
    words_with_offsets = tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str(text)
    new_words = [word for word, offset in words_with_offsets]
    for word in new_words:
        word_freqs[word] += 1

print(word_freqs)
````

````python
# base vocabulary
alphabet = set()

for word in word_freqs.keys():
    for letter in word:
        if letter not in alphabet:
            alphabet.add(letter)

alphabet = sorted(list(alphabet))
print(alphabet)
````

````python
vocab = ["<|endoftext|>"] + alphabet.copy()
print(vocab)
````

````python
# split all words into characters
splits = {word: [c for c in word] for word in word_freqs.keys()}
splits
````

````python
# compute character pair frequency
def compute_pair_freqs(splits):
    pair_freqs = defaultdict(int)
    for word, freq in word_freqs.items():
        split = splits[word]
        if len(split) == 1:
            continue
        for i in range(len(split)-1):
            pair = (split[i], split[i+1])
            pair_freqs[pair] += freq
    return pair_freqs

pair_freqs = compute_pair_freqs(splits)
````

````python
best_pair = ""
max_freq = None

for pair, freq in pair_freqs.items():
    if not max_freq or max_freq < freq:
        best_pair = pair
        max_freq = freq

print(best_pair, max_freq)
````

````python
merges = {('Ġ', 't'): "Ġt"}
vocab.append("Ġt")
````

````python
# merge the most frequent pair
def merge_pair(a, b, splits):
    for word in word_freqs:
        split = splits[word]
        if len(split) == 1:
            continue

        i = 0
        while i < len(split) - 1:
            if split[i] == a and split[i+1] == b:
                split = split[:i] + [a+b] + split[i+2:]
            else:
                i += 1
        splits[word] = split
    return splits

splits = merge_pair('Ġ', 't', splits)
````

````python
splits
````

Full loop

````python
vocab_size = 50

while len(vocab) < vocab_size:
    # find most frequent pair
    pair_freqs = compute_pair_freqs(splits)
    best_pair = ""
    max_freq = None
    for pair, freq in pair_freqs.items():
        if not max_freq or max_freq < freq:
            best_pair, max_freq = pair, freq

    # merge and get new splits
    splits = merge_pair(*best_pair, splits)
    merges[best_pair] = best_pair[0] + best_pair[1]
    vocab.append(best_pair[0] + best_pair[1])
````

# WordPiece tokenization

wordpiece prefix: `##`
- example: word --> `w ##o ##r ##d`


Merging by score: (prioritize less frequent ones)
```
score=(freq_of_pair)/(freq_of_first_element×freq_of_second_element)
```

Usage: find the longest subword in vocab, and split on it. If any part is unknown, the whole word will be tokenized as unknown.

### ***Implementation steps***
- corpus -> word_freqs -> vocabulary
- corpus **splits** -> find the best pair (by scores) -> **merge** the best pair -> add it to vocabulary
- REPEAT: splits and merge

````python
corpus = [
    "This is the Hugging Face Course.",
    "This chapter is about tokenization.",
    "This section shows several tokenizer algorithms.",
    "Hopefully, you will be able to understand how they are trained and generate tokens.",
]
````

````python
# pre-tokenize into words
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
````

````python
from collections import defaultdict

word_freqs = defaultdict(int)
for text in corpus:
    words_with_offsets = tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str(text)
    new_words = [word for word, offset in words_with_offsets]
    for word in new_words:
        word_freqs[word] += 1

word_freqs
````

````python
# vocabulary
alphabet = set()
for word in word_freqs.keys():
    if word[0] not in alphabet:
        alphabet.add(word[0])
    # other than the first letter, all others should have "##" prefix in WordPiece
    for letter in word[1:]:
        if f"##{letter}" not in alphabet:
            alphabet.add(f"##{letter}")

alphabet = list(alphabet)
alphabet.sort()
````

````python
vocab = ['[PAD]', '[UNK]', '[CLS]', '[SEP]', '[MASK]'] + alphabet.copy()
vocab
````

````python
splits = {
    word: [c if i == 0 else f'##{c}' for i, c in enumerate(word)]
    for word in word_freqs.keys()
}
splits
````

````python
def compute_pair_scores(splits):
    letter_freqs = defaultdict(int)
    pair_freqs = defaultdict(int)
    for word, freq in word_freqs.items():
        split = splits[word]
        if len(split) == 1:
            letter_freqs[split[0]] += freq
            continue
        for i in range(len(split) - 1):
            letter_freqs[split[i]] += freq
            pair_freqs[(split[i], split[i+1])] += freq
        letter_freqs[split[-1]] += freq  # don't forget the last element

    scores = {
        pair: pair_freq / (letter_freqs[pair[0]] * letter_freqs[pair[1]])
        for pair, pair_freq in pair_freqs.items()
    }
    return scores

pair_scores = compute_pair_scores(splits)
pair_scores
````

````python
best_pair = ""
max_score = float('-inf')
for pair, score in pair_scores.items():
    if score > max_score:
        best_pair, max_score = pair, score

print(best_pair, max_score)
````

````python
vocab.append('ab')
````

````python
# apply the merge in the `splits` dictionary
def merge_pair(a:str, b:str, splits:list):
    for word in word_freqs:
        split = splits[word]
        if len(split) == 1:
            continue
        i = 0
        while i < len(split) - 1:
            if split[i] == a and split[i+1] == b:
                merged = a+b[2:] if b.startswith('##') else a+b
                split = split[:i] + [merged] + split[i+2:]  # split is still a list
            else:
                i += 1
        splits[word] = split
    return splits
````

````python
splits = merge_pair('a', '##b', splits)
splits['about']
````

````python
# putting everything together
vocab_size = 70
while len(vocab) < vocab_size:
    pair_scores = compute_pair_scores(splits)
    # find best pair with highest score
    best_pair, max_score = None, float('-inf')
    for pair, score in pair_scores.items():
        if score > max_score:
            best_pair, max_score = pair, score
    # apply merge and generate new splits for the corpus
    splits = merge_pair(*best_pair, splits)
    new_token = (
        best_pair[0] + best_pair[1][2:]
        if best_pair[1].startswith('##')
        else best_pair[0] + best_pair[1]
    )
    vocab.append(new_token)
````

Use it to tokenize a new text

````python
def encode_word(word):
    tokens = []
    while len(word) > 0:
        if len(tokens) > 0:  # not the first token
            word = f"##{word}"
        j = len(word)
        # find the longest subword
        while j > 0 and word[:j] not in vocab:
            j -= 1
        if j == 0:  # no subword exists
            return ['[UNK]']
        tokens.append(word[:j])
        word = word[j:]  # start a new search from index j
    return tokens
````

````python
print(encode_word("Hugging"))
print(encode_word("HOgging"))
````

````python
def my_wordpiece_tokenize(text):
    pre_tokenized_result = tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str(text)
    pre_tokenized_text = [word for word, offset in pre_tokenized_result]
    encoded_words = [encode_word(word) for word in pre_tokenized_text]
    return sum(encoded_words, [])
````
